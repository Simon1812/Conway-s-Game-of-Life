```python
import yaml
import time 
from play import Game




config = yaml.safe_load(open("config.yaml", "r"))
config["sleep_duration"] = 0
config["grid_size"] = 200
config["max_steps"] = 100
config["visualization"]["mode"] = "none"


game = Game(config)
start_time = time.time()
game.start()
end_time = time.time()
print(f"Total time taken: {end_time - start_time} seconds")

# Starting Game of Life with grid size: (200, 200), max steps: 100, visualization mode: none
# Total time taken: 62.12204670906067 seconds
```
